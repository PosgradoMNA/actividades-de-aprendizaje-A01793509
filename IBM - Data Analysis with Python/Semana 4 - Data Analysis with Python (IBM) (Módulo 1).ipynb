```python
#Importamos la librer√≠a pandas y leemos el archivo auto.csv que se encuentra en una ruta p√∫blica de internet guardando los datos en un dataframe
import pandas as pd
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(url, header = None)
#Visualizamos los primeros 10 registros
df.head(10)
```

```python
#Ahora le asignaremos nombres a las columnas, para ello creamos una lista con los nombres de las columnas
CabColumnas = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
#Asignamos a las cabeceras de las columnas los nombres definidos
df.columns = CabColumnas
df.head(10)
```

```python
#Importamos la librer√≠a Numpy para realizar
import numpy as np

#Teniendo en cuenta que en la columna "normalized-losses" tenemos valores errados (?) los reemplazaremos con NAN (Not A Number)
df.replace('?',np.NaN)

#Teniendo en cuenta que en la columna "price" hay valores perdidos los eliminamos del conjunto
df.dropna(subset=["price"], axis=0)
```
